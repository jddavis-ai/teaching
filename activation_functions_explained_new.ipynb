# Activation Functions in Neural Networks

This notebook demonstrates several commonly used activation functions in neural networks. Each function is explained in detail, and applied to the same dataset so you can see how each one transforms the input data.

## What are Activation Functions?
Activation functions introduce non-linearity into a neural network, allowing it to learn complex patterns. Without them, a neural network would essentially be a linear model, no matter how many layers it has.

We'll visualize the following activation functions:
- ReLU (Rectified Linear Unit)
- Sigmoid
- Tanh (Hyperbolic Tangent)
- Leaky ReLU
- ELU (Exponential Linear Unit)
- Swish
- Softplus

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a global style for the plots
sns.set(style="whitegrid")
```

## Step 1: Create Input Data

We generate a sequence of values from -10 to 10 using `numpy.linspace`. This creates 400 evenly spaced values across that range. These values represent the input to each activation function.

```python
# Create input data
x = np.linspace(-10, 10, 400)
```

## Step 2: Define Activation Functions

Each function is defined as a Python function. Let's walk through them one by one:
- **ReLU:** Returns 0 for negative values and the value itself for positives.
- **Sigmoid:** Squashes values to a (0, 1) range. Good for binary classification.
- **Tanh:** Squashes values to a (-1, 1) range. Centered around zero.
- **Leaky ReLU:** Like ReLU but allows a small slope for negative values.
- **ELU:** Exponential behavior for negatives, linear for positives.
- **Swish:** A smooth function defined as `x * sigmoid(x)`. Proposed by Google.
- **Softplus:** A smooth version of ReLU. Always positive.

```python
# Define the activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

def softplus(x):
    return np.log(1 + np.exp(x))
```

## Step 3: Plot Activation Functions

We create a generic plotting function to show each activation function applied to the same data. This allows for direct visual comparison.

```python
# Function to plot activation function output
def plot_activation(x, y, title):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=title, linewidth=2)
    plt.title(f"{title} Activation Function")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.axvline(0, color='gray', linestyle='--', linewidth=1)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## Step 4: Visualize All Activation Functions

Now we apply each activation function to the input `x` and plot the results.

```python
# Dictionary of activation functions
activations = {
    "ReLU": relu,
    "Sigmoid": sigmoid,
    "Tanh": tanh,
    "Leaky ReLU": leaky_relu,
    "ELU": elu,
    "Swish": swish,
    "Softplus": softplus
}

# Plot each function
for name, func in activations.items():
    y = func(x)
    plot_activation(x, y, name)
```
